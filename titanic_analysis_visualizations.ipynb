# Análise de Sobrevivência no Titanic com Matplotlib

Neste notebook, analisaremos o dataset "Titanic Survival" usando a biblioteca `matplotlib`, com três visualizações diferentes e conclusões baseadas nos dados. O objetivo é entender os padrões de sobrevivência a partir de variáveis como gênero, idade e proporção geral de sobreviventes.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregando o dataset
df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

## 1. Gráfico de Barras: Sobrevivência por Gênero

```python
# Agrupando por sexo e sobrevivência
survival_by_sex = df.groupby(['Sex', 'Survived']).size().unstack()

# Plot
survival_by_sex.plot(kind='bar', stacked=True, color=['salmon', 'skyblue'])
plt.title('Sobrevivência por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Número de Pessoas')
plt.legend(['Não Sobreviveu', 'Sobreviveu'])
plt.tight_layout()
plt.show()
```

**Conclusão:**  
O gráfico mostra que mulheres tiveram uma taxa de sobrevivência significativamente maior do que os homens. Isso reflete a política de resgate “mulheres e crianças primeiro” adotada durante o desastre.

## 2. Histograma: Distribuição de Idade dos Passageiros

```python
# Histograma da idade
df['Age'].dropna().plot(kind='hist', bins=30, edgecolor='black')
plt.title('Distribuição de Idade dos Passageiros')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.tight_layout()
plt.show()
```

**Conclusão:**  
A maioria dos passageiros estava entre 20 e 40 anos. Também há uma quantidade considerável de crianças. A análise da idade pode ajudar a entender quais faixas etárias foram mais impactadas.

## 3. Gráfico de Pizza: Proporção de Sobreviventes

```python
# Contando os sobreviventes
survival_counts = df['Survived'].value_counts()

# Gráfico de pizza
plt.pie(survival_counts, labels=['Não Sobreviveu', 'Sobreviveu'], autopct='%1.1f%%', colors=['red', 'green'])
plt.title('Proporção de Sobreviventes')
plt.tight_layout()
plt.show()
```

**Conclusão:**  
Aproximadamente 38% dos passageiros sobreviveram, enquanto 62% não sobreviveram. Isso destaca a gravidade do naufrágio e a escassez de recursos para resgate.

## 4. Boxplot: Idade por Sobrevivência

```python
import seaborn as sns

# Boxplot idade x sobrevivência
plt.figure(figsize=(8, 5))
sns.boxplot(x='Survived', y='Age', data=df)
plt.xticks([0, 1], ['Não Sobreviveu', 'Sobreviveu'])
plt.title('Distribuição da Idade por Sobrevivência')
plt.xlabel('Sobreviveu')
plt.ylabel('Idade')
plt.tight_layout()
plt.show()
```

**Conclusão:**  
Passageiros que sobreviveram tendem a ser ligeiramente mais jovens. Isso sugere que idade pode ter influenciado nas chances de sobrevivência, possivelmente em conjunto com a prioridade dada a crianças durante o resgate.

## 5. Gráfico de Barras: Sobrevivência por Classe

```python
# Agrupando por classe e sobrevivência
survival_by_class = df.groupby(['Pclass', 'Survived']).size().unstack()

# Plot
survival_by_class.plot(kind='bar', stacked=True, color=['orange', 'lightgreen'])
plt.title('Sobrevivência por Classe do Embarque')
plt.xlabel('Classe')
plt.ylabel('Número de Pessoas')
plt.legend(['Não Sobreviveu', 'Sobreviveu'])
plt.tight_layout()
plt.show()
```

**Conclusão:**  
Passageiros da 1ª classe tiveram uma taxa de sobrevivência muito maior do que os da 3ª classe. Isso mostra que havia uma diferença de tratamento no acesso aos botes salva-vidas conforme a classe social a bordo.
